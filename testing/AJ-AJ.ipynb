```python
import pandas as pd
from tqdm._tqdm_notebook import tqdm_notebook
tqdm_notebook.pandas()
```

```python
with open('../2_1_Data_Acquisition_ORACC/output/oracc_akk.p', 'rb') as s: 
    word_df = pd.read_pickle(s)
```

```python
keep = ['lemma', 'pos', 'epos', 'sense', 'id_text', 'id_line', 'lang']
words = word_df[keep]
words
```

```python
ajaj = []
for i, p in tqdm_notebook(enumerate(words.pos)):
    if i == 0:
        continue
    if words.iloc[i].lang[:3] != "akk":
        continue
    if p == 'AJ' and words.iloc[i-1].pos == 'AJ':
        first = list(words.iloc[i-1])
        second = list(words.iloc[i])
        ajaj.append(first + second)
```

```python
ajaj_df = pd.DataFrame(ajaj)
```

```python
col = ['lemma', 'pos', 'epos', 'sense', 'id_text', 'id_line', 'lang', 'lemma2', 'pos2', 'epos2', 'sense2', 'id_text2', 'id_line2', 'lang2']
ajaj_df.columns = col
ajaj_df
```

```python
lines = word_df.groupby([word_df['id_text'], word_df['id_line'], word_df['label']]).agg({
        'form': ' '.join,
    }).reset_index()
lines
```

```python
ajaj_m = pd.merge(lines, ajaj_df, how='inner', on=['id_text', 'id_line'])
```

```python
ajaj_m = ajaj_m.rename(columns = {'form':'translit'})
```

```python
ajaj_m = ajaj_m[['id_text', 'label', 'translit', 'lemma', 'lemma2']]
```
