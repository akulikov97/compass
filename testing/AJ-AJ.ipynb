```python
import pandas as pd
```

```python
with open('../2_1_Data_Acquisition_ORACC/output/saao.p', 'rb') as s: 
    word_df = pd.read_pickle(s)
```

```python
keep = ['lemma', 'pos', 'epos', 'sense', 'id_text']
words = word_df[keep]
words
```

```python
ajaj = []
for i, p in enumerate(words.pos):
    if p == 'AJ' and words.iloc[i-1].pos == 'AJ':
        first = list(words.iloc[i-1])
        second = list(words.iloc[i])
        ajaj.append(first + second)
```

```python
ajaj_df = pd.DataFrame(ajaj)
```

```python
col = ['lemma', 'pos', 'epos', 'sense', 'id_text', 'lemma2', 'pos2', 'epos2', 'sense2', 'id_text2']
ajaj_df.columns = col
ajaj_df
```

```python
with open('ajaj.csv', 'w', encoding="utf-16") as a:
    ajaj_df.to_csv(a, index=False)
```

```python
with open('ajaj2.csv', 'w', encoding="utf-8") as a:
    ajaj_df.to_csv(a, index=False)
```
